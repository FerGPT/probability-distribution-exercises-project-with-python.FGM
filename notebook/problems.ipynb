## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment; for example, when you flip a coin, it can be either head or tails.

```python
print ("NORMAL DISTRIBUTION")

print ("\nLa diferencia entre la DISTRIBUCIÓN NORMAL y la DISTRIBUCIÓN BINOMIAL es que la normal es simétrica respecto a la media, y la distribució binomial no, estando desplazada en el eje X.")

#Por tanto, lo primero que tenemos que hacer es importar la librería corespondiente para la distribución normal. A continualción:

from scipy.stats import norm

# Define all the probability functions related to this distribution

def dnorm(x, mean=0, std_dev=1):
    """
    Calculates the probability density function of a normal distribution
    with the specified mean and standard deviation.
    """
    result = norm.pdf(x, loc=mean, scale=std_dev)

    return result

def pnorm(x, mean=0, std_dev=1):
    """
    Calculates the cumulative probability (CDF) of a normal distribution
    with the specified mean and standard deviation.
    """
    result = norm.cdf(x, loc=mean, scale=std_dev)
    return result

def qnorm(p, mean=0, std_dev=1):
    """
    Calculates the quantile (inverse CDF) for a normal distribution
    given a probability p, mean, and standard deviation.
    """
    result = norm.ppf(p, loc=mean, scale=std_dev)
    return result

def rnorm(n, mean=0, std_dev=1):
    """
    Generates random numbers from a normal distribution
    given the sample size, mean, and standard deviation.
    """
    result = norm.rvs(size = n, loc = mean, scale = std_dev)
    return result
```

```python
# Binomial(10, 0.2) distribution, where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial.
# Transformed into normal distribution.

import numpy as np

np.random.seed(42)

# Print statements for a normal distribution with appropriate arguments:
print(f"Probability density of a Normal(10, 0.2) at x=2: {dnorm(2, mean=10, std_dev=0.2)}")
print(f"Cumulative probability of a Normal(10, 0.2) at x=2: {pnorm(2, mean=10, std_dev=0.2)}")
print(f"Quantile for p=0.9 of a Normal(10, 0.2): {qnorm(0.9, mean=10, std_dev=0.2)}")
print(f"Generate 2000 random values from a Normal(10, 0.2): {rnorm(2000, mean=10, std_dev=0.2)}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)

mean = 0
std_dev = 1
data = np.arange(-3, 3, 0.01)

pmf = dnorm(data, mean = mean, std_dev = std_dev)
cdf = pnorm(data, mean = mean, std_dev = std_dev)
ppf = qnorm(data, mean = mean, std_dev = std_dev)

fig, axis = plt.subplots(3,1, figsize = (5, 7))

axis[0].plot(data, pmf, "k-")
axis[1].plot(data, cdf, "y-")
axis[2].plot(data, ppf, "m-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```
